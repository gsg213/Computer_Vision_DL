```
import numpy as np
import cv2
from matplotlib import pyplot as plt
```

```
img = cv2.imread("lena_noise.jpeg")
plt.imshow(img)
plt.show()
```

```
# gaussian kernel
kernel_gauss = np.array(([1,2,1],[2,4,2],[1,2,1]),dtype='float')*(1/16)
# box kernel
kernel_box =  np.array(([1,1,1],[1,1,1],[1,1,1]),dtype='float')*(1/9)
```

```
# box filter
img_box = cv2.filter2D(img,-1,kernel_box)
plt.imshow(img_box)
plt.show()
```

```
# gaussian filter
img_gauss = cv2.filter2D(img,-1,kernel_gauss)
plt.imshow(img_gauss)
plt.show()
```

```
#opencv gaussian blur
img_g = cv2.GaussianBlur(img,(3,3),0)
plt.imshow(img_g)
plt.show()
```

```
plt.subplot(131), plt.imshow(img_box), plt.title('Box filter')
plt.subplot(132), plt.imshow(img_gauss), plt.title('Gaussian filter')
plt.subplot(133), plt.imshow(img_g), plt.title('Opencv Gaussian Blur')
plt.show()
```
